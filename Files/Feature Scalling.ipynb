```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("loan_data_set.csv")
data.head(10)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.distplot(data["CoapplicantIncome"])
plt.show()
```

```python
ss = StandardScaler()
data[["CoapplicantIncome_ss"]] = ss.fit_transform(data[["CoapplicantIncome"]])
data.head(5)
```

```python
data.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(data["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(data["CoapplicantIncome_ss"])
plt.show()
```
