```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = pd.read_csv("relevant_sample_data.csv")
data
```

```python
data.isnull().sum() # Check Null Values
```

```python
data["Gender"].fillna(data["Gender"].mode()[0]) #Replace null Values
```

```python
data.isnull().sum() # After Fill null Value Recheck null Values
```

```python
en_data = data[["Gender"]]
en_data
```

```python
pd.get_dummies(en_data) # Using Pandas Perform encoding
```

# OneHotEncoding Using Scikit-Learn

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=["Female", "Male"])
```

# LabelEncoding Using Scikit-Learn

```python
data.isnull().sum()
```

```python
data["Department"].fillna(data["Department"].mode()[0])
```

```python
le_data = data["Department"]
```

```python
le = LabelEncoder()
data["Lable_Encode"] = le.fit_transform(le_data)
data[["Department","Lable_Encode"]]
```

# Ordinal Encoding

```python
df = pd.DataFrame({"Size":["l","s","m","l","xl","m","s"]})
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
oe = OrdinalEncoder(categories=ord_data)
df["Ordinal_Encoding"] = oe.fit_transform(df)
df
```

```python
# Ordinal Encoding Using map function
```

```python
ord_data1 = {"s":1,"m":2,"l":3,"xl":4}
df["Size_map"] = df["Size"].map(ord_data1)
df[["Size","Size_map"]]
```
